```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_path = "../data/raw/"
proccesed_path = "../data/processed/"
```

```python
df = pd.read_csv(os.path.join(raw_path, "clean2.data"), header=None)
```

```python
n_cols = df.shape[1]
columns = ["ID"] + [f"attr_{i}" for i in range(1, n_cols)]
df.columns = columns
```

Svaki red u clean2.data predstavlja jednu konformaciju molekula.
ID je naziv molekula dok su atributi koji ga prate numericke karakteristike molekula. Imamo 168 atributa po instanci koji predstavljaju razlicite karakteristike.
Kao sto vidimo prikazujuci head i tail imamo MUSK i NON-MUSK molekule. Naime, nas cilj je da klasifikujemo molekule, odnosno, da "predvidimo" da li je molekul musk ili non-musk. Pokusacemo da naucimo sta razlikuje musk i non-musk molekule.

```python
df.describe()
```

```python
numeric_df = df.select_dtypes(include='number') #skew i kurt za razliku od describe ne ignorisu ne-numericke kolone

print(numeric_df.skew()) #koliko je distribucija asimetricna
print(numeric_df.kurt()) #mjeri ostrinu distribucije, da li ima repove
```

```python
unique_ids = df['ID'].unique()
print(len(unique_ids))
```

```python
num_musk = sum(1 for x in unique_ids if x.startswith('MUSK'))
num_non_musk = sum(1 for x in unique_ids if x.startswith('NON-MUSK'))

print("Broj MUSK molekula:", num_musk)
print("Broj NON-MUSK molekula:", num_non_musk)
```

```python
import matplotlib.pyplot as plt
num_cols = df.columns[1:6] #prvih 5 numerickih atributa

df[num_cols].hist(figsize=(12,6), bins=20)
plt.suptitle("Histogram prvih 5 numeričkih atributa")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df['attr_1'], 
    y=df['attr_2'], 
    hue=df['ID'].apply(lambda x: 'MUSK' if x.startswith('MUSK') else 'NON-MUSK'),
    palette=['green','red']
)
plt.title("Scatter plot attr_1 vs attr_2 po klasi")
plt.show()
```

Ovaj scatterplot je jos uvijek informativnog karaktera i sluzi kao neka uvertira u samu pricu. X-osa je attr_1, Y-osa je attr_2 i svaka tacka na grafiku predstavlja jednu konformaciju sa svojim attr_1 i attr_2 vrijednostima.

```python
#Korelaciona matrica
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.iloc[:, :10].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Korelaciona matrica prvih 10 numeričkih kolona (automatski)")
plt.show()
```

Dakle vrijednost u celiji (i,j) predstavlja koeficijent korelacije izmedju atributa i sa atributom j.
Tumacenje vrijednosti:
+1 -> pozitivna linearna veza; ako jedan atribut raste i drugi raste proporcionalno
0  -> nema linearne zavisnosti
-1 -> negativna linearna veza; ako jedan atribut raste drugi opada proporcionalno

U ovom notebook-u započeli smo istraživanje dataset-a MUSK Version 2, koji se koristi za probleme klasifikacije u hemijskoj i bioinformatičkoj oblasti. Dataset sadrži 6598 konformacija molekula, od kojih svaka predstavlja jedinstvenu konfiguraciju molekula sa raznim fizičko-hemijskim karakteristikama. Svaka konformacija je opisana sa 168 numeričkih atributa koji kvantifikuju različite geometrijske i hemijske osobine, dok prvi atribut predstavlja ID molekula, koji nam omogućava da razdvojimo molekule koje su klasifikovani kao MUSK i one koje su NON-MUSK.
