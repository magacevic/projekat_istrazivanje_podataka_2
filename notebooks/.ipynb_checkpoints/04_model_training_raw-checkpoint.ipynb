```python
import pandas as pd
import numpy as np
import os
```

```python
raw_path = "../data/raw/"
```

```python
df = pd.read_csv(os.path.join(raw_path, "clean2.data"), header=None)
```

```python
n_cols = df.shape[1]
columns = ["ID"] + [f"attr_{i}" for i in range(1, n_cols)]
df.columns = columns
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_df = df.select_dtypes(include=np.number)

X_numeric = numeric_df.values
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_numeric)
```

```python
df['target'] = df['ID'].apply(lambda x: 1 if x.startswith('MUSK') else 0)
```

```python
X_scaled.shape
```

```python
X = X_scaled
y = df['target'].values

print("Numerički podaci skalirani i spremni za modele")
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2,       #koristimo 20% za test skup kao i prosli put
    random_state=42,     
    stratify=y           #koristimo stratify kao i prosli put
)
```

```python
print("Train/Test split gotov")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "DecisionTree": DecisionTreeClassifier(max_depth=10, min_samples_leaf=2),
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=42),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(n_neighbors=3)
}
```

Logisticka Regresija sa raw podacima

```python
lr_model = LogisticRegression(max_iter=1000)

lr_model.fit(X_train, y_train)

y_train_pred = lr_model.predict(X_train)
y_test_pred = lr_model.predict(X_test)

print("Logistic Regression - Train Accuracy:", accuracy_score(y_train, y_train_pred))
print("Logistic Regression - Test Accuracy: ", accuracy_score(y_test, y_test_pred))

print("\nClassification Report (Test set):")
print(classification_report(y_test, y_test_pred))

cm = confusion_matrix(y_test, y_test_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Confusion Matrix - LogisticRegression")
plt.show()
```

```python
import joblib
lr_model.fit(X, y)

joblib.dump(lr_model, "../notebooks/models/lr_raw_final.joblib")
print("Finalni Logistic Regression model (raw podaci) sačuvan kao lr_raw_final.joblib")
```

SVM sa raw podacima

```python
svm_model = SVC()

svm_model.fit(X_train, y_train)

y_train_pred = svm_model.predict(X_train)
y_test_pred = svm_model.predict(X_test)

print("SVM - Train Accuracy:", accuracy_score(y_train, y_train_pred))
print("SVM - Test Accuracy: ", accuracy_score(y_test, y_test_pred))

print("\nClassification Report (Test set):")
print(classification_report(y_test, y_test_pred))

cm = confusion_matrix(y_test, y_test_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Confusion Matrix - SVM")
plt.show()
```

```python
svm_model.fit(X, y)

joblib.dump(svm_model, "../notebooks/models/svm_raw_final.joblib")
print("Finalni SVM model (raw podaci) sačuvan kao svm_raw_final.joblib")
```
