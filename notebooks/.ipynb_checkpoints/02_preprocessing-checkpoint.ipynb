```python
import os
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
raw_path = "../data/raw/"
df = pd.read_csv(os.path.join(raw_path, "clean2.data"), header=None)

n_cols = df.shape[1]
columns = ["ID"] + [f"attr_{i}" for i in range(1, n_cols)]
df.columns = columns
df.head()
```

```python
#osnovni, brzi pregled podataka
#dimenzije
print("Dimenzije dataset-a:", df.shape)

#provjera nedostajucih vrijednosti
print("Nedostajući podaci po koloni:\n", df.isnull().sum())

#statistika numerickih atributa
print("Opis numeričkih kolona:\n", df.iloc[:,1:].describe())

#broj MUSK i NON-MUSK molekula
num_musk = sum(df['ID'].str.startswith('MUSK'))
num_non_musk = sum(df['ID'].str.startswith('NON-MUSK'))
print(f"MUSK: {num_musk}, NON-MUSK: {num_non_musk}")
```

```python
#Skaliranje podataka je potrebno kako bi PCA i modeli radili pravilno
#Prilikom skaliranja sve kolone postanu jednako vazne jer PCA trazi pravac najvece varijanse
#Naravno, skaliraju se samo numericki djelovi

numeric_df = df.select_dtypes(include=np.number) #moramo ovako da radimo jer u attr_1 imamo neke ne-numericke vrijednosti
X_numeric = numeric_df.values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
```

```python
#dodavanje kolone target u dataframe-u
df['target'] = df['ID'].apply(lambda x: 1 if x.startswith('MUSK') else 0)
```

```python
df[['ID', 'target']].head(10)
```

```python
df[['ID', 'target']].tail()
```

```python
n_components = 20
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
```

```python
print(pca.explained_variance_ratio_.sum()) #koliko ukupne informacije zadrzavamo u 20 komponenti, 86% varijansa
```

```python
pca_names = [f'PC{i+1}' for i in range(X_pca.shape[1])]
df_X_pca = pd.DataFrame(X_pca, columns=pca_names)
df_X_pca
```

```python
df_X_pca['target'] = df['target'].values
df_X_pca.head()
```

```python
df_X_pca.tail()
```

```python
print("Varijansa po komponenti:")
explained_vars = zip(pca_names, pca.explained_variance_ratio_)
for name, var in explained_vars:
    print(f"{name}: {round(var * 100, 2)}%")
```

```python
import matplotlib.pyplot as plt
cumsum_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(8,5))
plt.plot(range(1, n_components+1), cumsum_variance_ratio, marker='*', color='blue')
plt.xlabel('PCA component')
plt.ylabel('Cumulative explained variance')
plt.title('Cumulative explained variance by PCA components')
plt.grid(True)
plt.show()
```

Scatter plot PCA prikazuje strukturu podataka u prostoru sažetih komponenti. Kada koristimo PCA (Principal Component Analysis), originalni numerički atributi se transformišu u nove linearne kombinacije – principalne komponente – koje sažimaju što više varijanse iz podataka. Prve komponente (PC1, PC2, PC3) obično sadrže najveći dio informacije, pa je vizualizacija po ovim komponentama najinformativnija.

Na scatter plot-u, svaka tačka predstavlja jedan uzorak iz dataset-a, a boja označava ciljnu klasu (target). Na primjer, u mom Musk v2 dataset-u, crvena boja može predstavljati “non-Musk”, a zelena “Musk”.
Ako klase formiraju odvojene grupe u PCA prostoru, znači da su razlikujuće karakteristike prisutne u originalnim atributima.
Ako se klase preklapaju, znači da dataset nije lako separabilan, pa modeli imaju teže zadatke u klasifikaciji.

```python
plt.figure(figsize=(8,6))
for label, color, marker in zip([0,1], ['red','green'], ['o','^']):
    plt.scatter(
        df_X_pca[df_X_pca['target']==label]['PC1'],
        df_X_pca[df_X_pca['target']==label]['PC2'],
        label='MUSK' if label==1 else 'NON-MUSK',
        c=color,
        marker=marker,
        alpha=0.6
    )
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA scatterplot: PC1 vs PC2')
plt.legend()
plt.show()
```

### PCA – Principal Component Analysis

Za smanjenje dimenzionalnosti i eliminisanje medjukorelacija medju 168 numerickih atributa, koristila sam metodu **Principal Component Analysis (PCA)**.  
PCA transformise originalne atribute u novi skup komponenti koje su linearne kombinacije originalnih atributa, i pri cemu su komponente medjusobno nekorelisane.  

Odabrala sam **20 PCA komponenti**, sto objašnjava priblizno 86% ukupne varijanse originalnih podataka.  
Ovaj broj komponenti predstavlja kompromis između ocuvanja informacija i smanjenja kompleksnosti modela.

U dokumentu sam prikazala:  
1. DataFrame sa 20 PCA komponenti i kolonom target (MUSK(1)/ NON-MUSK(0))  
2. Varijansu koju objasnjava svaka komponenta  
3. Kumulativnu varijansu (graf)  
4. Scatterplot prve dvije PCA komponente radi vizualnog pregleda razdvajanja klasa  

```python
processed_path = "../data/processed"
df_X_pca.to_csv(os.path.join(processed_path, "clean2_pca.csv"), index=False)
```

```python
#eksperimentisanje
max_components = 40
pca_full = PCA(n_components=max_components)
X_pca_full = pca_full.fit_transform(X_scaled)
cumsum_var_full = np.cumsum(pca_full.explained_variance_ratio_)

#Brojevi komponenti koje zelimo prikazati
components_to_plot = [5, 20, 40]

#Izracunavamo kumulativnu varijansu za te komponente
cumulative_variances = [cumsum_var_full[n-1] for n in components_to_plot]

#Napravimo bar plot
plt.figure(figsize=(8,8))
plt.bar([str(n) for n in components_to_plot], cumulative_variances, color=['red', 'green', 'blue'])
plt.xlabel('Broj PCA komponenti')
plt.ylabel('Kumulativna objasnjena varijansa')
plt.title('Kumulativna varijansa za konkretan broj PCA komponenti')
plt.ylim(0, 1)  #varijansa je izmedju 0 i 1
for i, v in enumerate(cumulative_variances):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold')
plt.show()
```
